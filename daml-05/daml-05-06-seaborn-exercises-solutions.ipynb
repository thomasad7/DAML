```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = (12.5, 6.0)
import seaborn as sns
sns.set()
import numpy as np
import pandas as pd
tips = sns.load_dataset('tips')
tips.head()
```

### 1. Who pays better tips, smokers or non-smokers? (★☆☆)

```python
sns.distplot(tips[tips.smoker == 'No'].tip, bins=np.linspace(0, tips.tip.max(), 50))
sns.distplot(tips[tips.smoker == 'Yes'].tip, bins=np.linspace(0, tips.tip.max(), 50));
```

### 2. Are there specific days when smokers/non-smokers eat out?  During lunch or dinner? (★☆☆)

```python
sns.catplot('day', col='smoker', data=tips, hue='time', kind='count', height=6, aspect=1);
```

### 3. Which bill and tip values are most common?  Is there correlation between them? (★★☆)

```python
sns.jointplot(tips.total_bill, tips.tip, kind='kde', height=9);
```

### 4. Female or male waiters get bigger tips?  At which time of the day? (★★☆)

```python
grid = sns.FacetGrid(tips, row='sex', col='time', height=6)
grid.map(plt.hist, 'tip', bins=np.linspace(0, tips.tip.max(), 20));
```

### 5. When the number of meal courses increases, does the bill and tip increase accordingly? (★★☆)

```python
sns.jointplot(tips['size'], tips.total_bill, kind='reg', height=7)
sns.jointplot(tips['size'], tips.tip, kind='reg', height=7);
```

### 6. On which days and at what meals people take more courses? Being a smoker makes a difference? (★★☆)

```python
sns.catplot('size', row='time', col='smoker', data=tips, hue='day', kind='count', height=6, aspect=1);
```
